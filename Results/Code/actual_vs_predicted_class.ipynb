```python
import matplotlib.pyplot as plt
import numpy as np

cms = {
    'RF':       np.array([[154,15,6],[41,12,29],[19,7,97]]),
    'LR':       np.array([[132,37,6],[19,45,18],[6,27,90]]),
    'Ensemble': np.array([[142,27,6],[24,35,23],[8,22,93]])
}

labels = ['Home Win','Draw','Away Win']
x = np.arange(len(labels)) 
width = 0.35  

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
for ax, (model, cm) in zip(axes, cms.items()):
    actual_counts = cm.sum(axis=1) 
    predicted_counts = cm.sum(axis=0)

    ax.bar(x - width/2, actual_counts, width, label='Actual', color='#1f77b4')
    ax.bar(x + width/2, predicted_counts, width, label='Predicted', color='#ff7f0e')
    ax.set_title(model)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.set_ylabel('Number of Matches' if ax == axes[0] else "")
    ax.legend()

fig.suptitle('Actual vs Predicted Outcome Distribution by Model')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
